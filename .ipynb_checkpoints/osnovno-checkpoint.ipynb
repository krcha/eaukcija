```python
import requests
import json

# Setup for the request
url = "https://eaukcija.sud.rs/WebApi.Proxy/api/EAukcija/GetImmovablePropertyDetails"

headers = {
    "accept": "application/json",
    "accept-language": "en,en-US;q=0.9,sr;q=0.8,bs;q=0.7,hr;q=0.6,fi;q=0.5",
    "authorization": "bearer YOUR_ACCESS_TOKEN",  # You need to replace YOUR_ACCESS_TOKEN with your actual token
    "content-type": "application/json;charset=UTF-8",
}
data = {
    "AuctionId": "89432",
    "Role": None,
}

# Make the POST request
response = requests.post(url, headers=headers, json=data)
response_data = response.json()

# Function to recursively remove all Base64 fields from the response data
def remove_base64_fields(data):
    if isinstance(data, dict):
        return {key: remove_base64_fields(value) for key, value in data.items() if key.lower() != 'base64'}
    elif isinstance(data, list):
        return [remove_base64_fields(item) for item in data]
    else:
        return data

# Removing Base64 fields
cleaned_data = remove_base64_fields(response_data)

# Now you can save this cleaned data to a file or process it further as needed
# For example, saving to a JSON file without the Base64 fields
json_file_path = 'individual_pokretnosti_details.json'
with open(json_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(cleaned_data, json_file, ensure_ascii=False, indent=4)

print(f"Cleaned data saved to {json_file_path}")
```
